2- crie um algoritmo que leia 2 números e imprima na tela o resultado acima.

```python
número1=int(input("digite o primeiro número :"))
número2=int(input("digite o segundo número :"))


#processamento
soma=número1 + número2
subt=número1 - número2
div=número1 / número2
mult=número1 * número2
resto=número1 % número2
expo=número1 ** número2


#saída
print(f"a soma é: {soma} , e a  subtração é: {subt}")
print(f" a subtração é:{subt}")
print(f"A divisão é: {div:.2f}")
print(f" A multiplicação é: {mult}")
print(f" o resto da divisão é : {resto}")
print(f" A  exponenciação  é: {expo:.2e}")
```
